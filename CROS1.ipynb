# <center>Метод Розенброка<center>
# $$\frac{dx}{dt} = f(x, t)$$
# $$ \tau = (t_i - t_{i-1}) $$
# $$(E-\frac{1+i}{2}\tau f_x(x_{i-1}, t_{i-1}))k = f(x_{i-1}, t_{i-1} + 0.5\tau)$$
# $$x_i = x_{i-1} + \tau Re(k)$$

```python
import numpy as np
import scipy.sparse as ssp
import scipy.sparse.linalg as spl
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
plt.rc('font', size=20)
```

```python
def commutator(a, b):
    return a*b - b*a


def anti_commutator(a, b):
    return a*b + b*a


def jac_anti_commutator(a):
    return np.kron(a, np.eye(a.shape[0])) + np.kron(np.eye(a.shape[0]), a.T)


def lindbladian(*, A, gamma=0):
    return lambda ro: gamma * (A*ro*A.H - 0.5*anti_commutator(A.H*A, ro))


def jac_lindbladian(*, A, gamma=0):
    return gamma * (np.kron(A.H.T, A) - 0.5*jac_anti_commutator(A.H*A))


def fun(*, A, gamma=0):
    return lambda ro: lindbladian(A=A, gamma=gamma)(ro)


def jac(*, A, gamma=0):
    return jac_lindbladian(A=A, gamma=gamma)


def CROS1(*, ro_0, H, A, gamma, grid):
    F = lambda ro: np.array(fun(A=A, gamma=gamma)(ro)).ravel()

    a11 = (1+1j)/2

    step = grid[1] - grid[0]

    E = np.eye(ro_0.shape[0]*ro_0.shape[1])

    U = np.matrix(spl.expm(-1j*H*step))

    result = [ro_0]

    J = E - a11*step*jac(A=A, gamma=gamma)

    for _ in range(len(grid)-1):
        ro = result[-1]

        ro = U * ro * U.H

        k_imag = np.linalg.solve(J, F(ro.imag)).reshape(ro_0.shape).real
        k_real = np.linalg.solve(J, F(ro.real)).reshape(ro_0.shape).real

        ro = ro + step*(k_real + 1j*k_imag)

        result.append(ro)

    return result
```

## Пример - JC(1)

```python
w = 10**10
g = 10**8
gamma = 10**5
T = 10**-4
N = 10**3
grid = np.linspace(0, T, N)
```

```python
phi_0 = np.matrix([
    [0],
    [0],
    [1],
]).astype(np.complex128)

ro_0 = phi_0*phi_0.H

H = np.matrix([
    [0, 0, 0],
    [0, w, g],
    [0, g, w]
]).astype(np.complex128)

A = np.matrix([
    [0, 1, 0],
    [0, 0, 0],
    [0, 0, 0],
]).astype(np.complex128)
```

```python
%%time

ro = CROS1(
    ro_0=ro_0,
    H=H,
    A=A,
    gamma=gamma,
    grid=grid,
)
```

```python
ro00 = [ro[0,0].real for ro in ro]
ro11 = [ro[1,1].real for ro in ro]
ro22 = [ro[2,2].real for ro in ro]

plt.figure(figsize=(20,10))
plt.title("ro(t)")
plt.ylabel("ro")
plt.xlabel("t")
plt.plot(grid, ro00)
plt.plot(grid, ro11)
plt.plot(grid, ro22)
plt.grid()
plt.legend(["00", "11", "22"])
plt.show()
```

# Sparse

```python
def jac_commutator(a):
    return ssp.kron(a, ssp.eye(a.shape[0])) - ssp.kron(ssp.eye(a.shape[0]), a.T)


def jac_anti_commutator(a):
    return ssp.kron(a, ssp.eye(a.shape[0])) + ssp.kron(ssp.eye(a.shape[0]), a.T)


def jac_lindbladian(*, A, gamma=0):
    return gamma * (ssp.kron(A.H.T, A) - 0.5*jac_anti_commutator(A.H*A))


def jac(*, A, gamma=0):
    return jac_lindbladian(A=A, gamma=gamma)


def CROS1(*, ro_0, H, A, gamma, grid):
    ro_0 = ssp.csc_matrix(ro_0)
    H = ssp.csc_matrix(H)
    A = ssp.csc_matrix(A)

    F = (lambda ro: fun(A=A, gamma=gamma)(ro).toarray().ravel())

    a11 = (1+1j)/2

    step = grid[1] - grid[0]

    E = ssp.eye(ro_0.shape[0]*ro_0.shape[1])

    U = spl.expm(-1j*H*step)

    result = [ro_0]

    J = E - a11*step*jac(A=A, gamma=gamma)

    for _ in range(len(grid)-1):
        ro = result[-1]

        ro = U * ro * U.H

        k_imag = spl.spsolve(J, F(ro.imag)).reshape(ro_0.shape).real
        k_real = spl.spsolve(J, F(ro.real)).reshape(ro_0.shape).real

        k_real = ssp.csc_matrix(k_real)
        k_imag = ssp.csc_matrix(k_imag)

        ro = ro + step*(k_real + 1j*k_imag)

        result.append(ro)

    return [np.matrix(ro.toarray()) for ro in result]
```

## Пример - JC(n)

```python
w = 10**10
g = 10**8
gamma = 10**6
T = 10**-5
n = 201
N = 50
grid = np.linspace(0, T, N)
```

```python
phi_0 = np.matrix(np.zeros((n+1, 1))).astype(np.complex128)
phi_0[-1, -1] = 1
ro_0 = phi_0*phi_0.H


H = np.matrix(np.zeros((n+1, n+1))).astype(np.complex128)
H[range(n+1), range(n+1)] = w * (np.arange(n+1) // 2 + np.arange(n+1) % 2)
H[range(1, n+1), range(n+1-1)] = g * np.sqrt(np.arange(n+1-1))
H[range(n+1-1), range(1, n+1)] = g * np.sqrt(np.arange(n+1-1))

A = np.matrix(np.zeros((n+1, n+1))).astype(np.complex128)
A[range(n+1-1), range(1, n+1)] = np.sqrt(np.arange(1, n+1))
```

```python
%%time

ro = CROS1(
    ro_0=ro_0,
    H=H,
    A=A,
    gamma=gamma,
    grid=grid,
)
```

```python
plt.figure(figsize=(20,10))
plt.title("diagonal ro in the end")
plt.ylabel("diagonal ro")
plt.xlabel("i")
plt.plot(range(n+1), np.array(ro[-1]).diagonal().real)
plt.grid()
plt.show()
```
